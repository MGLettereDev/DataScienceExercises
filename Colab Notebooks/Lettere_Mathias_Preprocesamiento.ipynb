## Cuente cuantos valores faltantes tiene en cada variable.


```python
from pyspark.sql.functions import col, isnan, when, count, isnull
```

```python
#res = [count(when((col(c) == ' ')|( col(c).isNull()), c)).alias(c) for c in dfsPermitNew.columns]
#dfsPermitNew.select(*res).show()
def check_for_null_or_nan(df):
    null_or_nan = lambda x: isnan(x) | isnull(x)
    func = lambda x: df.filter(null_or_nan(x)).count()
    print(*[f'{i} has {func(i)} nans/nulls' for i in df.columns if func(i)!=0],sep='\n')
```

```python
check_for_null_or_nan(dfsPermit)
```

## Solvente los valores faltantes de cada variable. Aplique eliminación de filas y reemplazo de valores (valor promedio o categoría más frecuente)

```python
from pyspark.sql.functions import avg
```

```python
#Luego de un análisis de las variables y de los siguientes ejercicios, se determinó que las columnas de
# Number of Existing Stories y Estimated Cost seran llenadas con el valor promedio, Plansets y Revised Cost con 0 y Existing Units con la Mediana
# y el resto de columnas no necesitarán limpieza dado la cantidad de nulos y/o falta de relevancia para los siguientes ejercicios. Un nuevo DataFrame constará con los datos limpios

dfsTemp = dfsPermitNew.select(['Number of Existing Stories', 'Estimated Cost', 'Existing Units', 'Plansets', 'Revised Cost'])
avgNES = dfsTemp.agg(avg('Number of Existing Stories')).collect()[0][0]
avgEC = dfsTemp.agg(avg('Estimated Cost')).collect()[0][0]
avgEU = dfsTemp.approxQuantile('Existing Units', [0.5], 0.01)[0]

dfsClean = dfsTemp.fillna({'Number of Existing Stories': avgNES,  'Estimated Cost': avgEC, 'Existing Units': avgEU, 'Plansets': 0, 'Revised Cost': 0})
```

## Determine el valor máximo de la columna ‘Number of Existing Stories’.

```python
dfsClean.createOrReplaceTempView("PERMITS")
```

```python
spark.sql("SELECT MAX(`Number of Existing Stories`) AS Valor_Maximo FROM PERMITS").show()
```

## Determine el valor mínimo de la columna ‘Estimated Cost’.**bold text**

```python
spark.sql("SELECT MIN(`Estimated Cost`) AS Valor_Minimo FROM PERMITS").show()
```

## Determine la media de ‘Existing Units’.

```python
spark.sql("SELECT AVG(`Existing Units`) AS Promedio FROM PERMITS").show()
```

## Determine la mediana de ‘Plansets’.

```python
spark.sql("""
    SELECT
        percentile_approx(`Plansets`, 0.5) AS Mediana
    FROM
        PERMITS
""").show()
```

## Determine la desviación estándar de ‘Estimated Cost’. Interprete el resultado.

```python
spark.sql( """
    SELECT
        stddev(`Estimated Cost`) AS Desviacion_Estandar
    FROM
        PERMITS
""").show()
```

```python
spark.sql("SELECT MIN(`Estimated Cost`) AS Valor_Minimo, MAX(`Estimated Cost`) AS Valor_Maximo  FROM PERMITS").show()
```

Dado que la desviación es tan grande, vale observar los valores mínimo y máximos de la columna, de lo cual se puede encontrar que varian en gran medida, por lo tanto, en conjunto con la desviación estándar, se concluye que los datos son dispersos. Se puede inferir igualmente que aumentó la dispersión de los datos por lo que se utilizó la media para llenar los valores nulos. Es un ejemplo perfecto de por qué no siempre se debe llenar valores vacíos con la media.

## Determine el skewness de ‘Estimated Cost’. Interprete el resultado.

```python
spark.sql( """
    SELECT
        skewness(`Estimated Cost`) AS Skewness_valor
    FROM
        PERMITS
""").show()
```

Dado que este valor es mayor a 0, se observa que la simetría de los datos es asimétrica positiva.

## Determine la kurtosis de ‘Estimated Cost’. Interprete el resultado.

```python
spark.sql( """
    SELECT
        kurtosis(`Estimated Cost`) AS Kurtosis_valor
    FROM
        PERMITS
""").show()
```

```python
from pyspark.sql.functions import kurtosis

kurtosis_value = dfsClean.select(kurtosis(col('Estimated Cost'))).collect()[0][0]
print("El valor de la Kurtosis es: ", kurtosis_value)
```

Dado que el valor es mayor a 3, se concluye que la distribución de los datos es leptocúrtica. Principalmente debido a que los valores faltantes fueron llenados con la media, por lo tanto, se observa que no hay muchos valores considerados 'outliers'.

## Determine la correlación entre la variable ‘Estimated Cost’ y ‘Revised Cost. Interprete el resultado

```python
from pyspark.sql.functions import corr

corr_value = dfsClean.select(corr(col('Estimated Cost'), col('Revised Cost'))).collect()[0][0]

print(f"La correlación entre {'Estimated Cost'} y {'Revised Cost'} es: {corr_value}")
```

```python
spark.sql("""
    SELECT
        corr(`Estimated Cost`, `Revised Cost`) AS Correlacion
    FROM
        PERMITS
""").show()
```

Dada que la correlación es cercana a 1, se determina que ambas variables están fuertemente correlacionadas, por lo tanto en cualquier estudio seria mejor descartar una de ellas, dado que los resultados del mismo se podrian ver afectados negativamente si se incluye a ambas en la generación de un modelo.
