### Unificación del Dataframe

2.- Filtrar las columnas del DataFrame (ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc)

3.- Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
#Filtración
cols = ['ESTACION', 'MAGNITUD', 'ANO', 'MES']
cols.extend([col for col in dfEmisiones if col.startswith('D')])

dfEmReduced = dfEmisiones[cols]
dfEmReduced
```

```python
dfEmReduced = dfEmReduced.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')
dfEmReduced
```

4.-Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día

5.- Eliminar las filas con fechas no válidas y ordenar el DataFrame por estaciones contaminantes y fecha

```python
#Limpieza
dfEmReduced['DIA'] = dfEmReduced.DIA.str.strip('D')

#Concatenación de las columnas de ANO, MES y DIA
dfEmReduced['FECHA'] = dfEmReduced.ANO.apply(str) + '/' + dfEmReduced.MES.apply(str) + '/' + dfEmReduced.DIA.apply(str)

#Conversión a tipo de dato DATETIME
dfEmReduced['FECHA'] = pd.to_datetime(dfEmReduced.FECHA, format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')

dfEmReduced
```

```python
#Limpieza con isnat
dfEmReduced = dfEmReduced.drop(dfEmReduced[np.isnat(dfEmReduced.FECHA)].index)

#Ordenamiento del dataframe por estación, *contaminante y fecha
#Se asume que se refiere a al campo MAGNITUD, dado que es el único que expresaría el nivél de contaminación total, mas no su valor diario
dfEmReduced.sort_values(['ESTACION', 'MAGNITUD', 'FECHA'])
```

7.- Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
#Busca emisiones de cierta magnitud y en cierta estación en un rango de fechas
def emissions(estacion, magnitud, init, fin):
    return dfEmReduced[(dfEmReduced.ESTACION == estacion) & (dfEmReduced.MAGNITUD == magnitud) & (dfEmReduced.FECHA >= init) & (dfEmReduced.FECHA <= fin)].sort_values('FECHA').VALOR


emissions(35, 8, dt.datetime.strptime('2018/10/25', '%Y/%m/%d'), dt.datetime.strptime('2019/02/12', '%Y/%m/%d'))
```

10.- Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estación indicada.

```python
#Dado una magnitud y estación, devuelve un resumen de la emisión y contaminación
def contamSummary(estacion, magnitud):
    return dfEmReduced[(dfEmReduced.ESTACION == estacion) & (dfEmReduced.MAGNITUD == magnitud)].VALOR.describe()

print('Resumen #1:\n', contamSummary(35, 7),'\n', sep='')
print('Resumen #2:\n', contamSummary(54, 8),'\n', sep='')
print('Resumen #3:\n', contamSummary(58, 12), sep='')
```

11.- Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todos las estaciones.

```python
def contamSummaryMonthly(magnitud, ano):
    return dfEmReduced[(dfEmReduced.MAGNITUD == magnitud) & (dfEmReduced.ANO == ano)].groupby(['ESTACION', 'MES']).VALOR.mean().unstack('MES')

print('Resumen #1:\n', contamSummaryMonthly(9, 2019),'\n', sep='')
print('Resumen #2:\n', contamSummaryMonthly(5, 2016),'\n', sep='')
print('Resumen #3:\n', contamSummaryMonthly(8, 2017), sep='')
```

12.- Crear un función que reciba una estación de medición y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.

```python
def contamTypeSummaryMonthly(estacion, ano):
  data = dfEmReduced[(dfEmReduced.ESTACION == estacion) & (dfEmReduced.ANO == ano)].groupby(['MES']).VALOR.mean()
  df = pd.DataFrame(data)
  return df


contamTypeSummaryMonthly(18, 2019)
```
